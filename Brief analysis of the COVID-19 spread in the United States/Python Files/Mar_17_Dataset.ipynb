### Single Linear Regression
To find the attributes that have relation to the number of cases.
The coefficient and pvalue is found out.

```python
# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

# create a fitted model in one line
Pop_Cases_lm = smf.ols(formula='Cases_Mar_17 ~ Population', data=data).fit()

# print the coefficients
print(Pop_Cases_lm.params)
print(Pop_Cases_lm.pvalues)
```

### Multiple Linear Regression
